#### נגדיר שיטת מדידה לביצועי המודלים שלנו

RMSE נשתמש במדד השגיאה הריבועית הממוצעת

הסיבה לכך היא שמדדים כמו דיוק לא מייצגים בצורה טובה מספיק את המציאות. 

**דוגמה להמחשה**: אם הציון הנכון של משתמש על מוצר הוא 2 ומודל ראשון חזה ציון 3 ושני ציון 5, ברור כי המודל הראשון מוצלח יותר, אך במדדי דיוק שני המודלים שגו באותו האופן

RMSEנשתמש בממוצע השגיאות המרובעות, ונבצע עבורו שורש בעצמנו לטובת קבלת ה

```python
# working on lists as well
def rmse(y_true, y_pred):
    return np.sqrt(mean_squared_error(y_true, y_pred))
```

```python
# example sets:
ratings_true = [1, 5, 2, 4, 1, 3]
ratings_pred = [2, 3, 2, 2, 1, 4.5]

rmse(ratings_true, ratings_pred)
```

#### evaluating a model on our test data

נגדיר מודל בסיסי שאליו נוכל לבצע השוואה עם המודלים שלנו. מטרתו - להבין כמה שיפור ביצענו בעזרת האלגוריתם. כל מודל שנגדיר יקבל שני פרמטרים - מספר משתמש ומספר מוצר, והוא יצטרך להחזיר כפלט את הציון החזוי לאותו המוצר.

הסיבה שאנו בונים את המודלים בצורה הזו, היא כדי לפתור את בעיה מספר 2 אותה תיארנו למעלה ושאותה אנחנו מעוניינים לפתור. נעתיק אותה בשנית:

**בהינתן משתמש ומוצר, על המערכת לחזות את הציון (דירוג) שהמשתמש היה נותן לאותו המוצר**

---

נבנה מודל בסיס שיחזה את הציון הממוצע לכל סרט מבין הטווח 0.5 ועד 5, שהוא 2.75 

```python
# we will define working models as a function that can get a single user and an item, and produce a rating
# this is exactly problem no.2 defined above

# baseline model for example
def baseline_model(user, item):
    # return middle score between 0.5 and 5
    return 2.75
```

נגדיר פונ' עזר שמחשבת את הציון של האלגוריתמים שלנו. מה שהיא עושה בפועל זה לקחת את נתוני הטסט שלנו, לבצע חיזוי לכל משתמש וסרט שיש לנו עליו נתון - ואז להשוות RMSE-לציון המקורי שנתן אותו המשתמש, ולגזור מההפרש את ערך ה 

```python
# define a score function for evaluation of the rmse on a model
def score(model):
    
    user_movie_pairs = zip(X_test['userId'], X_test['movieId'])
    
    y_pred = np.array([model(user, movie) for (user, movie) in user_movie_pairs]).reshape(1, -1)
        
    y_true = np.array([X_test['rating']])
    
    return rmse(y_true=y_true, y_pred=y_pred)
```

לדוג', להלן מימוש של פונקציית הציון על המודל הבסיסי שהגדרנו:

```python
# example of model usage
score(baseline_model) # this is the score that we would like to beat with our models
```

#### Training models

כעת נבצע הרצה של שני אלגוריתמים של מערכות המלצה כדי לבצע השוואה לתוצאות:

1. הראשון מבוסס למידה בלתי מפוקחת של ביצוע קליסטור של המשתמשים לפי הדירוגים שלהם, ומתן הציוג לפי ממוצע מבוסס המשתמשים הדומים להם בלבד לפי תוצאת הקלסיטור
2. השני מבוסס תהליך כפל מטריצות שנועד לחלץ מס' משתנים מצומצם לכל משתמש ולכל מוצר. מבוסס על פתרון לתחרות של נטפליקס משנת 2006 של סימון פאנק

```python
# import the surprise recommender system library
from surprise import KNNBasic, SVD
from surprise.model_selection import cross_validate
from surprise import Reader, Dataset
```

```python
knn_alg = KNNBasic(k=100, min_k=1)

svd_alg = SVD()
```

reader נטען את הנתונים לפי מה שדורשת הספרייה - לתוך אובייקט שנקרא

נשים לב כי בשלב זה אנו משתמשים באלגוריתמים אלו אך ורק בעזרת נתוני הדירוג של המשתמשים. אנו לא משתמשים כלל בנתוני מסגרת נוספים שנתקלנו בהם בשלב האקספלורציה במידע. דוגמה לנתונים נוספים שיכולנו להשתמש בהם בתהליך:
1. התיוגים שנתנו המשתמשים על סרטים
2. הז'אנרים של הסרטים השונים

```python
# use a reader to parse the csv dataFrame
reader = Reader(rating_scale=(1, 5))

data = Dataset.load_from_df(X_train[['userId', 'movieId', 'rating']], reader)
```

---

וכעת, לאחר טעינת הנתונים נבצע הרצות של שני האלגוריתמים בנפרד:

ונתעלם מנתוני המבחן בשלב זה KFOLD נחדד כי בהרצה הנוכחית נשתמש בחלוקת הנתונים ובאימון בשיטת  

#### KNN הרצת

```python
print(cross_validate(knn_alg,
               data=data,
               measures=['rmse'],
               cv=2))
```

#### SVD הרצת

```python
print(cross_validate(svd_alg,
                     data=data,
                     measures=['rmse'],
                     cv=2))
```

2. עם סט נתונים זה לא נדרשנו כמעט בכלל לבצע ניקוי או בדיקה מעמיקה של הנתונים (נתונים חסרים, נתונים מחוץ לטווח, וכו'). יש לממש בהמשך הפרויקט תהליך שעושה ניקוי מסודר לנתונים - משמיט ערכים לא תקינים, משלים/משמיט ערכים חסרים, וכו'.
3. מימוש אלגוריתמים של שיטות היברידיות שמשלבים גם פיצ'רים על המוצרים / המשתמשים עצמם (כדוגמת הז'אנרים של הסרטים או התגיות של המשתמשים בסט הנתונים הנ"ל), ולשלב את הנתונים האלו יחד עם שימוש באלגוריתמים שבוחנים רק את דירוגי המשתמשים.
4. מימוש השוואות מסודרות ובחינת מטריקות נוספות להשוואה בין הנתונים. לדוגמא:

MAE, Maybe Recall + Precision? More matrices for Ordinal Regression

5. חלוקה נוספת של הנתונים גם לסט של ולידיישן. משום שאנחנו רוצים לבחון מודלים שונים ופרמטרים שונים, אנחנו מקבלים מצב שההערכה שלנו על סט הטסט איננה מייצגת מספיק. נרצה לבצע חלוקה נוספת של הדאטה ל-3 חלקים בפועל, לאמן על חלק האימון, לבחור מודל לפי התוצאות על הולידיישן, ואז להשתמש בטסט בשביל הערכת השגיאה האמיתית של המודל שלנו
6. במטרה להסתכל על הפרויקט כמשימה של דאטה סיינטיסט, אני מעוניין "להלביש" על גבי המודל מתודה שתנגיש את השימוש בפלט שלו. לדוג', הנגשה ויזואלית של דירוגים ממוצעים של המנוע את המשתמשים השונים, ממשק שמאפשר למשתמש להכניס מוצר ולקבל את הדירוג של המנוע עליו / המלצות של המנוע עבור אותו המשתמש, וכדומה. דורש מחקר לגבי ויזואליזציות והנגשות רלוונטיות בעולם של מערכות המלצה.

**הערה לגבי סעיף 3**: יש לבצע מחקר לגבי יכולת השימוש בפיצ'רים על המוצרים / המשתמשים עצמם, שכן האלגוריתמים הקיימים לא יודעים להשתמש בצורה אקטיבית בנתוני "מסגרת" נוספים על המשתמשים ועל המוצרים.
